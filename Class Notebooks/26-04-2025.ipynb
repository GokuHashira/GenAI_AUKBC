```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from sklearn.neighbors import KernelDensity

# Define the desired distribution to sample from:
d = 2  # Number of dimensions
mean = np.array([[0.], [1.]]).flatten()
covariance = np.array([
    [1, 0.8],
    [0.8, 1]
])

# Compute the Decomposition:
A = np.linalg.cholesky(covariance)

# Sample X from standard normal
n = 50  # Samples to draw
Z = np.random.normal(size=(d, n))

# Apply the transformation
Y = A.dot(Z) + mean.reshape(-1, 1)
Y = Y.T  # Transpose for easier handling by KernelDensity

# Estimate the density using Kernel Density Estimation (KDE)
kde = KernelDensity(kernel='gaussian', bandwidth=0.5)  # Adjust bandwidth as needed
kde.fit(Y)
log_density = kde.score_samples(Y)
density = np.exp(log_density)

# Plot the samples with color based on density
fig, ax = plt.subplots(figsize=(6, 4.5))
scatter = ax.scatter(Y[:, 0], Y[:, 1], c=density, cmap='viridis', alpha=0.6,
                   edgecolor='k', linewidth=0.5)
ax.set_xlabel('y1', fontsize=13)
ax.set_ylabel('y2', fontsize=13)
ax.axis([-2.5, 2.5, -1.5, 3.5])
ax.set_aspect('equal')
ax.set_title('Samples from bivariate normal distribution (Density Colored)')
cbar = plt.colorbar(scatter)
cbar.ax.set_ylabel('Sample Density', fontsize=13)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from sklearn.neighbors import KernelDensity

# Define the desired distribution to sample from:
d = 2  # Number of dimensions
mean = np.array([[0.], [1.]]).flatten()
covariance = np.array([
    [1, 0],
    [0, 1]
])

# Compute the Decomposition:
A = np.linalg.cholesky(covariance)

# Sample X from standard normal
n = 50  # Samples to draw
Z = np.random.normal(size=(d, n))

# Apply the transformation
Y = A.dot(Z) + mean.reshape(-1, 1)
Y = Y.T  # Transpose for easier handling by KernelDensity

# Estimate the density using Kernel Density Estimation (KDE)
kde = KernelDensity(kernel='gaussian', bandwidth=0.5)  # Adjust bandwidth as needed
kde.fit(Y)
log_density = kde.score_samples(Y)
density = np.exp(log_density)

# Plot the samples with color based on density
fig, ax = plt.subplots(figsize=(6, 4.5))
scatter = ax.scatter(Y[:, 0], Y[:, 1], c=density, cmap='viridis', alpha=0.6,
                   edgecolor='k', linewidth=0.5)
ax.set_xlabel('y1', fontsize=13)
ax.set_ylabel('y2', fontsize=13)
ax.axis([-2.5, 2.5, -1.5, 3.5])
ax.set_aspect('equal')
ax.set_title('Samples from bivariate normal distribution (Density Colored)')
cbar = plt.colorbar(scatter)
cbar.ax.set_ylabel('Sample Density', fontsize=13)
plt.show()
```
